```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt


dados = pd.read_csv('base.csv').to_dict(orient='records')
df = pd.DataFrame(dados)
```

### Arvore ID3

```python
X = df[["Histria de Crdito", "Renda"]]
y = df["Risco"]

X_enc = pd.get_dummies(X)

# Treina modelo
clf = DecisionTreeClassifier(criterion="entropy", random_state=42)
clf.fit(X_enc, y)

plt.figure(figsize=(8,6))
plot_tree(
    clf,
    feature_names=X_enc.columns,
    class_names=clf.classes_,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.show()
```

## rvore de Deciso C4.5 

```python
X_c45 = df[["Renda", "Histria de Crdito", "Garantia"]]
y_c45 = df["Risco"]

X_c45_enc = pd.get_dummies(X_c45)

# Treinar rvore 
clf_c45 = DecisionTreeClassifier(
    criterion="entropy",
    min_samples_leaf=1,
    ccp_alpha=0.01,
    random_state=42
)
clf_c45.fit(X_c45_enc, y_c45)

plt.figure(figsize=(10,6))
plot_tree(
    clf_c45,
    feature_names=X_c45_enc.columns,
    class_names=clf_c45.classes_,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.show()
```

### rvore de Deciso CART

```python
dados = pd.read_csv('base.csv').to_dict(orient='records')
df = pd.DataFrame(dados)

X_cart = df[["Histria de Crdito", "Renda"]]
y_cart = df["Risco"]

X_cart_enc = pd.get_dummies(X_cart)

# Treinar rvore 
clf_cart = DecisionTreeClassifier(
    criterion="gini",
    random_state=42
)
clf_cart.fit(X_cart_enc, y_cart)

plt.figure(figsize=(8,6))
plot_tree(
    clf_cart,
    feature_names=X_cart_enc.columns,
    class_names=clf_cart.classes_,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.show()
```
